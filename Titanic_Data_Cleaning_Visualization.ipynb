```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')

df.head()
```

```python
df.info()

df.describe()

df.shape
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)

df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

df.drop('Cabin', axis=1, inplace=True)
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(8,4))

sns.boxplot(x=df['Fare'])

plt.title('Fare Outliers')

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Survived', data=df)

plt.title('Passenger Survival')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Sex',
    hue='Survived',
    data=df
)

plt.title('Gender vs Survival')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Pclass',
    hue='Survived',
    data=df
)

plt.title('Class vs Survival')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Age'],
    kde=True,
    bins=30
)

plt.title('Age Distribution')

plt.show()
```

```python
plt.figure(figsize=(10,6))

corr = df.select_dtypes(include='number').corr()

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

Key Insights

1. Females had significantly higher survival rates.
2. First-class passengers had better chances of survival.
3. Most passengers were adults aged 20-40.
4. Fare prices varied significantly with passenger class.
5. Passenger class influenced survival more than age.
